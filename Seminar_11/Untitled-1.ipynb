```python
import math
from sympy import *
from sympy.plotting import plot
# from numpy import *

x = symbols('x')
fx = -12*x**4 - 18*x**3+5*x**2 + 10*x - 30
df = diff(fx)
print(f'корни {fx}')
print(solve(fx))
print(f'дифференциал {fx}')
print(df)
print('функция возрастает')
print(solve(df > 0))
print('функция убывает')
print(solve(df < 0))
print('вычисление вершин')
print(solve(df))
print('f > 0')
print(solve(fx > 0))
print('f < 0')
print(solve(fx < 0))
print('печать графика')
plot(fx)
```
